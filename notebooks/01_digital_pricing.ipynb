```python
import numpy as np
import plotly.graph_objects as go
from exotics.engine import ExoticOptions
```

```python
r, q, sigma = 0.02, 0.0, 0.20
K = 100.0
T_ST= 0.01
```

```python
S_grid = np.linspace(80, 120, 200)
option = ExoticOptions(S_grid, r, q, sigma)
```

```python
price_digital_call = option.price_digital_call(K, T_ST)
price_replication_widespread = option.price_digital_replication(K, T_ST, 2)
price_replication_tightspread = option.price_digital_replication(K, T_ST, 0.1)
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=S_grid, y=price_digital_call, mode='lines', 
                         name='Theoretical Price', line=dict(color='black', width=3, dash='dash')))

fig.add_trace(go.Scatter(x=S_grid, y=price_replication_widespread, mode='lines', 
                         name='Call Wide Spread  - eps=2.0', line=dict(color='blue', width=2)))

fig.add_trace(go.Scatter(x=S_grid, y=price_replication_tightspread, mode='lines', 
                         name='Call Tight Spread - eps=0.1', line=dict(color='red', width=2)))

fig.add_vline(x=K, line_width=1, line_dash="dot", line_color="grey")

fig.update_layout(
    title="<b>Pricing Digital: Theoretical vs Call Spread Replication</b><br>",
    xaxis_title="Spot Price (S)",
    yaxis_title="Option Value",
    template="plotly_white",
    hovermode="x unified"
)

fig.show()
```
