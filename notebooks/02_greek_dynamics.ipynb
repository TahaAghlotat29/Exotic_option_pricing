```python
import numpy as np
import plotly.graph_objects as go
from exotics.engine import ExoticOptions
from exotics.greeks import Greeks
import kaleido
```

```python
S_grid = np.linspace(90, 110, 400)
engine = ExoticOptions(S_grid, r=0.02, q=0.0, sigma=0.2)
greeks = Greeks(engine)
K = 100.0
```

```python
maturities = {'1 Year': 1.0, '1 Month': 1/12, '1 Day': 1/252}
```

```python
fig_delta = go.Figure()

for label, T in maturities.items():
    d = greeks.digital_delta(K, T)
    fig_delta.add_trace(go.Scatter(x=S_grid, y=d, name=f'Delta ({label})'))

fig_delta.update_layout(
    title="<b>Digital Delta </b><br>",
    xaxis_title="Spot", yaxis_title="Delta",
    template="plotly_white", hovermode="x"
)
fig_delta.write_image("../images/digital_delta.png", scale=2 , width=1000, height=400)
```

```python
T_short = 0.01

fig_gamma = go.Figure()

gamma = greeks.digital_gamma(K, T_short)

fig_gamma.add_trace(go.Scatter(x=S_grid, y=gamma, name='Gamma', line=dict(color='red')))
fig_gamma.add_vline(x=K, line_dash="dash", line_color="black")
fig_gamma.add_hline(y=0, line_color="grey")

fig_gamma.update_layout(
    title="<b>Digital Gamma - T = 0.01 </b><br>",
    xaxis_title="Spot", yaxis_title="Gamma",
    template="plotly_white"
)
fig_gamma.write_image("../images/digital_gamma.png", scale=2 , width=1000, height=400)
```

```python
fig_vega = go.Figure()

vega = greeks.digital_vega(K, T=0.01)

fig_vega.add_trace(go.Scatter(x=S_grid, y=vega, name='Vega', line=dict(color='red', width=3)))
fig_vega.add_vline(x=K, line_dash="dash", line_color="black")
fig_vega.add_hline(y=0, line_color="grey")

fig_vega.update_layout(
    title="<b>Digital Vega </b><br>",
    xaxis_title="Spot", yaxis_title="Vega",
    template="plotly_white"
)
fig_vega.write_image("../images/digital_vega.png", scale=2 , width=1000, height=400)
```
